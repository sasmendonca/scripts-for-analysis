```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import MACCSkeys
from rdkit.DataStructs import BulkTanimotoSimilarity

# 1. Carregar o arquivo CSV
input_file = 'C:/Users/sabri/OneDrive/Documentos/UFG/Similarity/input/NTS_josue.csv' 
df = pd.read_csv(input_file)

# Certifique-se de que a coluna com os SMILES se chama 'SMILES'
# Se o nome for diferente, ajuste abaixo
smiles_column = 'SMILES'

# 2. Compostos específicos para calcular a similaridade
compostos_especificos = {
    'a': 'O=C1C(C(C(O)=C2C)=C3C(O)=C2NC(/C=C\C=C\C(C)C4O)=O)=C(C(C)=C3O)OC1(C)O/C=C/C(O)C(C)C(OC(C)=O)C(C)C(O)C4C', 
    'b': 'C[C@]([C@@H](CC)OC([C@H](C)C([C@@H]1C)=O)=O)(O2)[C@@H]([C@@H](C)C([C@H](C)C[C@](OC/C=C/C3=CC4=CC=C(C)N=C4C=C3)(C)[C@@H]1O[C@H]5[C@H](O)[C@@H](N(C)C)C[C@@H](C)O5)=O)NC2=O', 
    'c': 'O=C(/C=C/1)[C@H](CCCCCCCCC[C@@H](C)OC1=O)OC(CCC(O)=O)=O', 
    'd': 'C[C@H](CC(/C=C(C)/[C@@H](C)C[C@H](O)CC([C@H](O)[C@H]1O)=O)=C)C[C@H](O2)[C@@H]2/C=C/CC/C=C(C)/C(O[C@H](O)CC[C@H]1C)=O', 
    'e': 'C[C@H](CC(/C=C(C)/[C@@H](C)C[C@H](O)CC([C@@H]1O)=O)=C)C[C@H]2[C@H](/C=C/CC/C=C(C)/C(O[C@@H](CO)C[C@@H](C)[C@@H]1O)=O)O2', 
    'f': 'O=C(C1)OC(C/C=C/C=C(C)/C(OP(O)(O)=O)CCCC/C=C/C=C/C=C/C(C)C1=C)CC/C(C)=C/C=C', 
    'g': 'OC1=C(C(O[C@@H](C)CCC[C@@H]2O)=O)C(/C=C/C[C@@H](O)C2=O)=C(OC)C(OC)=C1',
    'h': 'CC1C2CC(O[C@H]1/C=C/CC3O[C@@H]4C[C@@H](O)C3C)[C@H](/C=C/C=C\C=C/C(OC(C=C5)C(/C(C)=C/C(C)CCCCC(O)=O)OC5C/C=C/CC/C=C/C(O)[C@H]4O)=O)O2'
}

# 3. Calcular fingerprints MACCS para todos os compostos
df['Mol'] = df[smiles_column].apply(Chem.MolFromSmiles) 
df['MACCS'] = df['Mol'].apply(MACCSkeys.GenMACCSKeys)

# 4. Calcular o coeficiente de Tanimoto para cada composto específico
for nome, smiles in compostos_especificos.items():
    # Converte o SMILES do composto específico para molécula e fingerprint
    mol_especifico = Chem.MolFromSmiles(smiles)
    if mol_especifico is None:
        print(f"Erro: SMILES inválido para {nome} ({smiles})")
        continue
    maccs_especifico = MACCSkeys.GenMACCSKeys(mol_especifico)
    
    # Calcula a similaridade de Tanimoto entre o composto específico e todos os outros
    df[f'Tanimoto_{nome}'] = df['MACCS'].apply(lambda x: BulkTanimotoSimilarity(maccs_especifico, [x])[0])

# 5. Remover colunas temporárias (se não forem necessárias)
df.drop(columns=['Mol', 'MACCS'], inplace=True)

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# 6. Salvar o DataFrame atualizado em um novo arquivo CSV
output_file = 'C:/Users/sabri/OneDrive/Documentos/UFG/Similarity/output/similaridade_macrolideos_2.csv'
df.to_csv(output_file, index=False)

print(f"Arquivo salvo com sucesso: {output_file}")
```
